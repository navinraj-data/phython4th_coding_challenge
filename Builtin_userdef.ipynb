```python
print (' Day 45 – Built-in Functions and User-Defined Functions')

# Define basic arithmetic functions
print('1. Create functions for addition, subtraction, multiplication, and division.')
def add(x, y):
  return x + y

def sub(x, y):
  return x - y

def mul(x, y):
  return x * y

def div(x, y):
  if y != 0:
    return x / y
  else:
    return "Cannot divide by zero"
print()
print('2.Ask the user to input two numbers and the operation they want to perform.')
firstnum= float(input('enter your first number: '))
secondnum=float(input('enter your second number: '))
operator=input('enter any operator +, -,*,/:' )
print()
print('3.Call the appropriate function and display the result.')
if operator == '+':
  result = add(firstnum, secondnum)
elif operator == '-':
  result = sub(firstnum, secondnum)
elif operator == '*':
  result= mul(firstnum,secondnum)
elif operator == '/':
  result=div(firstnum,secondnum)
else :
  result='invalid operation'
  print(result)

# Add this line to print the result for valid operations as well
if result != 'invalid operation':
  print(f"Result: {result}")
  print()
  print('4.Use built-in functions like abs(), round(), and pow() within your solution.')
  print(f"Absolute value of {firstnum}: {abs(firstnum)}")
  print(f'round of {secondnum}: {round(secondnum)}')
  print(f'{firstnum} to the power of {secondnum}: {round(pow(firstnum,secondnum))}')
  print('Final result is :', result)



```
