```python
print('Shopping Cart System')

total_items = 0
subtotal = 0

while True:
    price = float(input("Enter item price, 0 to stop response): "))

    if price == 0:
        break

    if price < 0:
        print(" Negative prices are not allowed. please Try again.")
        continue

    total_items += 1
    subtotal += price


if subtotal >= 1000:
    discount = subtotal * 0.20
elif subtotal >= 500:
    discount = subtotal * 0.10
elif subtotal >= 200:
    discount = subtotal * 0.05
else:
    discount = 0

final_total = subtotal - discount


print("\n--- BILL SUMMARY ---")
print("Total items:", total_items)
print("Subtotal:", subtotal)
print("Discount:", discount)
print("Final Total:", final_total)
```
